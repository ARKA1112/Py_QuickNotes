```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats
from statsmodels.tsa.arima.model import ARIMA
```

```python
from statsmodels.graphics.api import qqplot
```

```python
# get the sunspots data

print(sm.datasets.sunspots.NOTE)
```

```python
data = sm.datasets.sunspots.load_pandas().data
```

```python
data
```

```python
data.index = pd.Index(sm.tsa.datetools.dates_from_range("1700", "2008"))
data.index.freq = data.index.inferred_freq
```

```python
data
```

```python
# now lets delete the YEAR

del data['YEAR']
data.plot(figsize=(12,8))
```

```python
fig  = plt.figure(figsize=(12,9))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(data.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(data, lags=40, ax=ax2)
```

```python
arma_mod20 = ARIMA(data, order=(2,0,0)).fit()
print(arma_mod20.params)
```

```python
arma_mod30 = ARIMA(data, order=(3,0,0)).fit()
print(arma_mod20.aic, arma_mod20.bic, arma_mod20.hqic)
```

```python
print(arma_mod30.params)
```

```python
sm.stats.durbin_watson(arma_mod30.resid.values)
```

```python
#plotting the residuals
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
ax = arma_mod30.resid.plot(ax=ax)
```

```python
resid = arma_mod30.resid
```

```python
stats.normaltest(resid)
```

```python
fig = plt.figure(figsize=(12,9))
ax = plt.subplot(111)
fig = qqplot(resid, line='q', ax=ax, fit=True)
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid, lags=40, ax=ax2)
```

```python
r, q, p = sm.tsa.acf(resid.values.squeeze(), fft=True, qstat=True)
data = np.c_[np.arange(1, 25), r[1:], q, p]
```

```python
table = pd.DataFrame(data, columns=["lag", "AC", "Q", "Prob(>Q)"])
print(table.set_index("lag"))
```

```python
predict_sunspots = arma_mod30.predict("1990","2012", dynamic=True)
print(predict_sunspots)
```

```python
# check the error

def mean_forecast_err(y,yhat):
    return y.sub(yhat).mean()
```

```python
mean_forecast_err(data.SUNACTIVITY, predict_sunspots)
```

```python
from statsmodels.tsa.arima_process import ArmaProcess
```

```python
np.random.seed(1232)
# include zero-th lag
arparams = np.array([1, 0.75, -0.65, 0.9])
maparams = np.array([1, 0.65])
```

```python
# Lets make sure this model is estimable


arma_t = ArmaProcess(arparams, maparams)
arma_t.isinvertible
```

```python
arma_t.isstationary
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
ax.plot(arma_t.generate_sample(nsample=50))
```
